```python
from bs4 import BeautifulSoup
import json
from datetime import datetime
from tqdm import tqdm
#import torch
```

```python
data = None
with open("horoscopes.txt") as fp:
    data = json.load(fp)
```

```python
cleaned = []

sign_map = {
    1:"Aries",
    2:"Taurus",
    3:"Gemini",
    4:"Cancer",
    5:"Leo",
    6:"Virgo",
    7:"Libra",
    8:"Scorpio",
    9:"Sagittarius",
    10:"Capricorn",
    11:"Aquarius",
    12:"Pisces",
}

for key in tqdm(data):
    content = BeautifulSoup(data[key])
    horoscope = content.p.contents[1][3:]

    sign = sign_map[int(key.split(":")[1])]
    date = datetime.strptime(key.split(":")[0], '%Y%m%d').strftime("%B %d, %Y")
    
    cleaned.append(f"{sign} for {date}: {horoscope}")

```
